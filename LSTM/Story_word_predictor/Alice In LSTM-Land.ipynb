```python
data = reduce(lambda x,y: x+y,lines)
data = data.replace('\n',' ')
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
pre_data=pd.DataFrame({'x1':range(0,1000),'x2':range(0,1000),'x3':range(0,1000),'x4':range(0,1000),'x5':range(0,1000),'y':range(0,1000)})
```

```python
X=pre_data.iloc[:755,:5]
```

```python
vocab_size=len(tokenizer.word_index)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.25)
```

```python
train_X = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))
test_X = X_test.values.reshape((X_test.shape[0], 1, X_test.shape[1]))
print(train_X.shape, y_train.shape, test_X.shape, y_test.shape)
```

```python
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(vocab_size, activation='sigmoid'))
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')
```

```python
model.fit(train_X, y_train, epochs=500, verbose=2,validation_data=(test_X, y_test))
```

```python
def get_int(value):
    m_str = str(value)
    return int(m_str.split('.')[0])
```

```python
def get_word_key(value):
    int_value = get_int(value)
    for k, v in tokenizer.word_index.items():
        if v == int_value:
            return k
    return "none"
```
